思考一下我們前面有提到圖片是矩陣，但維度可能會不一樣
例如灰階圖只有兩個維度，RGB 彩圖則有 3 個維度

假如今天我們把 RGB 3 個維度拆開來看會有甚麼不同的效果呢？

Answer:
分別呈現出三種顏色(RGB)的圖片

```python
!pip install --user opencv-python

%matplotlib inline
from matplotlib import pyplot as plt
```

```python
import cv2

img_path = 'data/lena.png'

# 以彩色圖片的方式載入
img = cv2.imread(img_path, cv2.IMREAD_COLOR)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
# 以灰階圖片的方式載入
img_gray = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
plt.imshow(cv2.cvtColor(img_gray, cv2.COLOR_BGR2RGB))
```

```python
imgR = cv2.imread(img_path, cv2.IMREAD_COLOR)
imgR[:,:,0] = 0
imgR[:,:,1] = 0

plt.imshow(cv2.cvtColor(imgR, cv2.COLOR_BGR2RGB))
```

```python
imgG = cv2.imread(img_path, cv2.IMREAD_COLOR)
imgG[:,:,0] = 0
imgG[:,:,2] = 0

plt.imshow(cv2.cvtColor(imgG, cv2.COLOR_BGR2RGB))
```

```python
imgB = cv2.imread(img_path, cv2.IMREAD_COLOR)
imgB[:,:,1] = 0
imgB[:,:,2] = 0

plt.imshow(cv2.cvtColor(imgB, cv2.COLOR_BGR2RGB))
```
